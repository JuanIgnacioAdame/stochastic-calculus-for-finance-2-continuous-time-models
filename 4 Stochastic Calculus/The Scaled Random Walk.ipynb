```python
from matplotlib import pyplot as plt
from random import choice
```

```python
M = [0]
# n, number of coin tosses per unit time
n = 1000
# t, amount of (unitless) time you want to carry out the simulation
t = 10
# 
x = [0]
for _ in range(1, t*n+1):
    x.append(1/n*_)
    value = M[-1] + 1/(n**0.5)*choice([-1,1])
    M.append(value)
#print(M)
```

```python
plt.plot(x, M)
plt.show()
```
