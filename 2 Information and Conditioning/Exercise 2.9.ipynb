### Exercise 2.9

Let $X$ be a random variable.

### Exercise 2.9.1

Give an example of a probability space $(\Omega,\mathcal{F},\mathbb{P})$, a random variable $X$ defined on this probability space, and a function $f$ so that the $\sigma$-algebra generated by $f(X)$ is not the trivial $\sigma$-algebra $\{\emptyset,\Omega\}$ but is strictly smaller than the $\sigma$-algebra generated by $X$.

Edit: The textbook does not mention this, but it might be necessary that the function $f$ be Borel measurable.  I believe this is just a technical detail that we need so that it is guaranteed that some things be well-defined.  Eventually I will have to confirm this.

### Solution 2.9.1

Let $X:(\mathbb{R},\mathcal{B}(\mathbb{R}),\mathbb{P})\rightarrow (\mathbb{R},\sigma(X),\mu_X)$ be standard normal.  One way to construct such a random variable is to define $\mathbb{P}$ by
\begin{equation}
\forall A\in\mathcal{B}(\mathbb{R}), \quad \mathbb{P}(A)=\int_{A}\varphi(x)dx,
\end{equation}
where $\varphi(x)$ denotes the standard normal density.  We then define $X(\omega)=\omega$.  In this case, we have $\sigma(X)=\mathcal{B}(\mathbb{R})$, and $\mu_X=\mathbb{P}$.

Let $f(X)=|X|$.  We want to show that $\sigma(f(X))\subsetneq \sigma(X)$.  First we show that for any $A\in\sigma(f(X))$, then $A\in\sigma(X)$.  Then we give an example of some $A\in\sigma(X)$ such that $A\not\in\sigma(f(X))$.

Let $A\in\sigma(f(X))$.  Then, for some Borel set $B\in\mathcal{B}(\mathbb{R})$, we can write $A$ as
\begin{gather}
A=\{\omega\in\mathbb{R}: f(X(\omega))\in B\}=\{\omega\in\mathbb{R}: |\underbrace{X(\omega)}_{=\omega}|\in B\}=\{\omega\in\mathbb{R}: |\omega|\in B\}=B\cup(-B)\\
=\{X(\omega)=\omega\in B\}\cup\{X(\omega)=\omega\in-B\}.
\end{gather}
where $-B=\{-\omega:\omega\in B\}$.  Based on the way every set above is written, it is immediately clear that they are in $\sigma(X)$.  In short, for the arbitrary $A\in\sigma(f(X))$ is also holds that $A\in\sigma(X)$.

Note, for example, that $A=\{\omega:X(\omega)=\omega\in[0,\infty)\}=[0,\infty)$ is *not* in $\sigma(f(X))$.  The reason for this is that for any set $A'\in\sigma(f(X))$, it holds that for every $x\in A'$, then $-x\in A'$, which is a property that $A$ does *not* have.  To see this, note that for any $A'\in\sigma(f(X))$, then by definition (and also the computations we did above) there is a $B\in\mathcal{B}(\mathbb{R})$ such that $A'=\{\omega\in\mathbb{R}: f(X(\omega))\in B\}=\{X(\omega)=\omega\in B\}\cup\{X(\omega)=\omega\in-B\}$.  In short, $A\not\in\sigma(f(X))$.

Putting everything above together, we see that $\sigma(f(X))\subsetneq \sigma(X)$. $\square$

Remark: I *believe* the proof above works for any function $f$ that is surjective and not injective.  Such a function "loses" information because by mapping two different points to the same point, you can no longer know what the original point was.

### Exercise 2.9.2

Can the $\sigma$-algebra generated by $f(X)$ ever be strictly larger than the $\sigma$-algebra generated by $X$?

### Solution 2.9.2

This is not possible. Intuitively, this is because "knowing" $X(\omega)$, you immediately "know" $f(X(\omega))$ by merely plugging in the value $X(\omega)$ into the known function $f$.

More rigorously, we have
\begin{equation}
\sigma(f(X))\ni A=\{\omega\in\mathbb{R}: f(X(\omega))\in B\}=\{\omega\in\mathbb{R}: X(\omega)\in f^{-1}(B)\}\in\sigma(X).
\end{equation}
Although $f^{-1}$ is not well-defined in general, the pre-image of $B$ under $f$, which we denote $f^{-1}(B)$, is.  This establishes the result. $\square$
