### Exercise 1.4

### Exercise 1.4.1

Construct a standard normal random variable $Z$ on the probability space $(\Omega_{\infty}, \mathcal{F}_{\infty},\mathbb{P})$ of Example 1.1.4 under the assumption that the probability for head is $p=\frac{1}{2}$. (Hint: consider Examples 1.2.5 and 1.2.6)

### Solution 1.4.1

The main idea will be to combine the results of Example 1.2.5 and Example 1.2.6.

We start as in Example 1.2.5.  For $n=1,2,\dots$, define $Y_n$ by
\begin{align}
Y_n(\omega)&=1 \text{ if } \omega_n=H,\\
&=0 \text{ if } \omega_n=T,
\end{align}
and let $X$ be defined by
\begin{gather}
X:(\Omega_{\infty}, \mathcal{F}_{\infty}, \mathbb{P})\rightarrow([0,1], \mathcal{B}[0,1], \mu_X)\\
X(\omega)=\sum_{n=1}^{\infty}\frac{Y_n(\omega)}{2^n}.
\end{gather}
Then $X$ thus defined is a random variable defined on the infinite coin toss space, and uniformly distributed in $[0,1]$. 

Now we borrow from Example 1.2.6. Let $N^{-1}$ be the inverse of the function $N$, which is defined as
\begin{align}
&N:\mathbb{R}\rightarrow[0,1],\\
&N(x)=\int_{-\infty}^{x}\varphi(\xi)d\xi.
\end{align}

As show in Example 1.2.6, the random variable $Z:=N^{-1}\circ X$ is a standard normal random variable on the probability space $(\Omega_{\infty}, \mathcal{F}_{\infty}, \mathbb{P})$.  In other words,
\begin{align}
&Z:(\Omega_{\infty}, \mathcal{F}_{\infty}, \mathbb{P})\rightarrow(\mathbb{R}, \mathcal{B}(\mathbb{R}), \mu_X),\\
&\mu_Z[a,b]=\int_{a}^{b}\varphi(x)dx. \square
\end{align}

### Exercise 1.4.2

Define a sequence of random variables $\{Z_n\}_{n=1}^{\infty}$ on $\Omega_{\infty}$ such that
\begin{equation}
\lim_{n\rightarrow\infty}Z_n(\omega)=Z(\omega) \text{ for every } \omega\in\Omega_{\infty}
\end{equation}
and, for each $n$, $Z_n$ depends only on the first $n$ coin tosses. (This gives us a procedure for approximating a standard normal random variable by random variables generated by a finite number of coin tosses, a useful algorithm for Monte Carlo simulation).

### Solution 1.4.2

Such a sequence is given by
\begin{equation}
Z_n:=N^{-1}\circ\sum_{i=1}^{n}\frac{Y_i}{2^i}.
\end{equation}

To see this, we carry out the following calculations.  First,
\begin{align}
\lim_{n\rightarrow\infty}Z_n(\omega)&=\lim_{n\rightarrow\infty}N^{-1}\left(\sum_{i=1}^{n}\frac{Y_i(\omega)}{2^i}\right)\\
&=N^{-1}\left(\lim_{n\rightarrow\infty}\sum_{i=1}^{n}\frac{Y_i(\omega)}{2^i}\right)\\
&=N^{-1}(X(\omega))=Z(\omega).
\end{align}
Note for that second equality, we were allowed to bring the limit inside the function because $N^{-1}$ is continuous.  This establishes the result. $\square$
